At this point, we are addressing the calculation of the Flight Hunter Index metric. To achieve this, we need to develop a precise function for calculating the distance between two geographical points while considering the Earth's oblate spheroid shape. To accomplish this, we utilized the geopy.distance library, specifically its built-in geodesic function, which aligns with our requirements.

```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from datetime import datetime
from geopy.distance import geodesic
# Load the data from CSV files into pandas DataFrames
df = pd.read_csv('C:\\Users\\gothu\\Downloads\\FinalSQLforPython.csv')
df2 = pd.read_csv('C:\\Users\\gothu\\Downloads\\rawdata1.csv')


# Function to calculate distance between two geographical points considering Earth as an oblate spheroid
def calculate_distance(row):
    coords_1 = (row['home_airport_lat'], row['home_airport_lon'])
    coords_2 = (row['destination_airport_lat'], row['destination_airport_lon'])
    # Check for missing or NaN values in coordinates
    if any(pd.isna(val) for val in coords_1) or any(pd.isna(val) for val in coords_2):
        return None
    # Use the geodesic function to calculate the distance in kilometers
    return geodesic(coords_1, coords_2).kilometers

# Apply the distance calculation function to the DataFrame df2
df2['distance'] = df2.apply(calculate_distance, axis=1)

# Function to calculate Average Discount per Kilometer (ADS/KM) for each user group
def calculate_ads_per_km(group):
    total_discount_amount = (group['fd_amount'] * group['base_fare_usd']).sum()
    total_distance = group['distance'].sum()
    # Check to avoid division by zero
    if total_distance == 0:
        return 0
    return total_discount_amount / total_distance

# Group data by 'user_id' and calculate ADS/KM for each group
grouped = df2.groupby('user_id').apply(calculate_ads_per_km).reset_index(name='ads_per_km')

# Min-Max Scaling of ADS/KM values to bring them within [0, 1] range
min_ads_per_km = grouped['ads_per_km'].min()
max_ads_per_km = grouped['ads_per_km'].max()
grouped['scaled_ads_per_km'] = (grouped['ads_per_km'] - min_ads_per_km) / (max_ads_per_km - min_ads_per_km)

# Merge the original DataFrame df with the grouped DataFrame on 'user_id'
df = pd.merge(df, grouped[['user_id', 'scaled_ads_per_km']], on='user_id', how='left')

# Compute the 'flight_hunter_index' as a weighted combination of several features
# This index aims to identify users who are primarily looking for flight discounts
df['flight_hunter_index'] = df['scaled_ads_per_km'] * df['flight_discount_proportion'] * df['average_flight_discount']

# Print the column names for final DataFrame (useful for debugging or review)
print(df.columns)

# Save the modified DataFrame to a new CSV file
df.to_csv('C:\\Users\\gothu\\Downloads\\correctOblateData.csv', index=False)
```

## Metrics Used for Segmentation

### Travel Preferences (Flights, Hotels or Both)

Metric/s: ‘prefer_hotel’, ‘prefer_flight’, ‘prefer_both’.

Description: Knowing if the customer likes to book flights, hotels, or a combination helps us send them offers that actually interest them.

Please refer to the graphic below for a visualisation of the User Preferences across all “Traveler Tribes”.



```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from a CSV file into a pandas DataFrame
df = pd.read_csv('C:\\Users\\gothu\\Downloads\\FinalFulldata.csv')

# Visualize User Preferences by Cluster

# Get the unique cluster IDs, sorted
unique_clusters = np.sort(df['cluster'].unique())

# Initialize lists to store the sum of preferences for flights, hotels, and both
prefers_flights_list = []
prefers_hotels_list = []
prefers_both_list = []

# Loop over each unique cluster to calculate the sum of each type of preference
for cluster in unique_clusters:
    filtered_df = df[df['cluster'] == cluster]
    prefers_flights_list.append(filtered_df['prefers_flights'].sum())
    prefers_hotels_list.append(filtered_df['prefers_hotels'].sum())
    prefers_both_list.append(filtered_df['prefers_both'].sum())

# Number of unique clusters
n_clusters = len(unique_clusters)

# Define bar chart parameters
ind = np.arange(n_clusters) 
width = 0.20       

# Create the bar chart for preferences
plt.figure(figsize=(8.27,5.87))
bar1 = plt.bar(ind, prefers_flights_list, width, label='Prefers Flights')
bar2 = plt.bar(ind + width, prefers_hotels_list, width, label='Prefers Hotels')
bar3 = plt.bar(ind + 2 * width, prefers_both_list, width, color='purple', label='Prefers Both')

custom_labels= ['Discount Enthusiasts', 'Balanced Travellers', 'Hotel Discount Lovers', 'Jet-Set & Indecisive', 'Hotel Loyalists']

# Add some text for labels, title and axes ticks
plt.ylabel('User Preference Metric by Tribe')
plt.title('User Preferences Across Tribes')
plt.xticks(ind + width, custom_labels, rotation=45, ha='right')
plt.legend()

# Save the Image with 1080p resolution
plt.savefig('D:\\TravelTide\\userpreferences.png', dpi=96, bbox_inches='tight')

# Show the bar chart
plt.show()
```

### Age and User Count Distribution

Description: The customer age can tell us a lot about the kind of travel experiences they might be interested in, how often they might travel and what budget they might have.

Please refer to the graphic below for a visualisation of the Users and Average Age across all “Traveler Tribes”.

```python
# Create Users and Average Age graphic from all Clusters.
# Calculate the total count of users per cluster
total_users_per_cluster = df.groupby('cluster')['age'].count()

# Calculate the average age for each cluster
average_age_per_cluster = df.groupby('cluster')['age'].mean()

# Get the unique cluster IDs
unique_clusters = np.sort(df['cluster'].unique())

# Define bar chart parameters
n_clusters = len(unique_clusters)
ind = np.arange(n_clusters) 
width = 0.20
bar_width = 0.4

# Create the figure and the first y-axis for total users (bar1)
fig, ax1 = plt.subplots(figsize=(8.27,5.87))

bar1 = ax1.bar(np.arange(n_clusters), total_users_per_cluster, width, label='Total Users per Tribe')

# Set y-axis label and title for the first y-axis
ax1.set_ylabel('Total Users per Tribes')
ax1.set_title('Total Users and Average Age per Tribes')

# Create a second y-axis for average age (bar2)
ax2 = ax1.twinx()

bar2 = ax2.bar(np.arange(n_clusters) + width, average_age_per_cluster, width, label='Average Age per Tribe', color='orange')

# Set y-axis label for the second y-axis
ax2.set_ylabel('Average Age per Tribe')

custom_labels= ['Discount Enthusiasts', 'Balanced Travellers', 'Hotel Discount Lovers', 'Jet-Set & Indecisive', 'Hotel Loyalists']

# Add some text for x-axis labels, legend, and axes ticks
ax1.set_xticks(ind + width)
ax1.set_xticklabels(custom_labels, rotation=45, ha='right')
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Save the Image with 1080p resolution
plt.savefig('D:\\TravelTide\\users_and_average_age.png', dpi=96, bbox_inches='tight')

# Show the bar chart
plt.show()
```

### Ease of Booking

Metric/s: ‘conversion_rate’, ‘average_clicks’.

Description: Fewer clicks to make a booking means they had an easier time on the platform, and we want to keep it that way. Could also mean they found what they wanted fairly quickly such as discounts.

Please refer to the graphic below for a visualisation of the Engagement & Efficiency across all “Traveler Tribes”.

```python
# Aggregate conversion_rate and average_clicks by cluster and reset the index
# 'conversion_rate' is summed up, and 'average_clicks' is averaged for each cluster.
cluster_summary = df.groupby('cluster').agg({
    'conversion_rate': 'sum',
    'average_clicks': 'mean'
}).reset_index()

# Sort the DataFrame by the 'cluster' column
cluster_summary = cluster_summary.sort_values(by='cluster')

# Define bar chart parameters
bar_width = 0.5


# Create a new figure and a set of subplots (just one in this case)
fig, ax1 = plt.subplots(figsize=(8.27,5.87))

custom_labels= ['Discount Enthusiasts', 'Balanced Travellers', 'Hotel Discount Lovers', 'Jet-Set & Indecisive', 'Hotel Loyalists']

# Create a bar graph for 'conversion_rate' on the first set of axes (ax1)
# This represents Engagement
ax1.bar(cluster_summary['cluster'], cluster_summary['conversion_rate'], width=bar_width, color='b', alpha=0.6, label='Total Conversions')
ax1.set_xticks(cluster_summary['cluster'])
ax1.set_xticklabels(custom_labels, rotation=45, ha='right')
ax1.set_ylabel('Total Conversions', color='b') # Set the y-axis label
ax1.tick_params('y', colors='b') # Set the color for the y-axis tick parameters

# Create a second set of axes sharing the same x-axis (ax2)
ax2 = ax1.twinx()

# Create a line graph for 'average_clicks' on the second set of axes (ax2)
# This represents Efficiency
ax2.plot(cluster_summary['cluster'], cluster_summary['average_clicks'], color='orange', label='Average Clicks')
ax2.set_ylabel('Average Clicks', color='orange')  # Set the y-axis label
ax2.tick_params('y', colors='orange') # Set the color for the y-axis tick parameters

# Combine legends from both ax1 and ax2 and display them
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Set the overall title for the graph
plt.title('Engagement & Efficiency by Tribe')

# Save the Image with 1080p resolution
plt.savefig('D:\\TravelTide\\engagementandefficiency.png', dpi=96, bbox_inches='tight')

# Display the graph
plt.show()
```
