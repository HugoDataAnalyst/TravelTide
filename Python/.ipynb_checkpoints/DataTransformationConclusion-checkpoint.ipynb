# Clustering our Dataset
To cluster the metrics introduced in the Introduction and Analysis & Findings sections, we employed the k-means clustering algorithm within a Jupyter Notebook. K-means is an unsupervised machine learning model that enables us to effectively segment our users based on the aforementioned metrics.

By incorporating these aggregated metrics and leveraging advanced analytics, we gain a deeper understanding of customer behauvior. This understanding, in turn, allows us to tailor perks to meet the specific needs and desires of each "traveller tribe."

This approach ensures that the perks not only reflect the unique attributes of each customer "tribe" but also align with their travel patterns and preferences. Final Data Transform
Before creaing theseting these customer categories, it was essential to ensure that all data were equally weighted to generate accurate insights.

In the following code, I took several steps to ensure data equalization:

- Calculated the age of each user efficiently and expressed it as a numerical value.
- Transformed the 'latest_session' time data into a more usable format.
- Utilized the StandardScaler function, which enables data to be scaled to a Z-Score standardization.
- Printed the values assigned to as its execution.

```python
# Import required libraries for data manipulation, feature scaling, clustering, and visualization.
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from datetime import datetime, timedelta
import random
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset. Make sure to specify the path correctly.
df = pd.read_csv('C:\\Users\\gothu\\Downloads\\correctOblateData2.csv')

# Compute the 'age' feature from 'birthdate'
# I use the floor division operator to get the integer value of age.
current_date = datetime.now()
df['birthdate'] = pd.to_datetime(df['birthdate'])
df['age'] = (datetime.now() - df['birthdate']).dt.total_seconds() // (365.25 * 24 * 3600)

# Obtain the average age of the dataset to use as reference.
mean_age = df['age'].mean()
print(f"Mean Age: {mean_age:.2f}")
# Convert 'latest_session' to datetime and calculate 'days_since_last_session'
# Useful for capturing recent behaviour.
df['latest_session'] = pd.to_datetime(df['latest_session'])
df['days_since_last_session'] = (current_date - df['latest_session']).dt.days

# Feature scaling: important due to differing units among features.
# First, MinMax scaling for algorithms that are sensitive to magnitudes.
# MinMax scales features to a range [0, 1].
features_to_scale = ['total_cancellation_rate', 'average_checked_bags', 'conversion_rate',
                     'prefers_flights', 'prefers_hotels', 'prefers_both',  
                     'discount_responsiveness', 'flight_discount_proportion','average_flight_discount',
                     'hotel_discount_proportion', 'average_hotel_discount', 'both_discount_proportion',
                     'flight_hunter_index', 'hotel_hunter_index', 
                     'total_hotel_usd_spent', 'total_flight_usd_spent', 'age'
                    ]

# Create new column names for scaled features
scaled_columns = ["scaled_" + feature  for feature in features_to_scale]
minmaxscaler = MinMaxScaler()
df[scaled_columns] = minmaxscaler.fit_transform(df[features_to_scale])


# Second, Z-score scaling (StandardScaler) to transform features to have mean=0 and variance=1.
# Suitable for algorithms that assume features are centered around zero.
zscaled_columns = [feature + "_zscaled" for feature in features_to_scale]
scaler = StandardScaler()
df[zscaled_columns] = scaler.fit_transform(df[features_to_scale])

# Apply K-Means clustering. n_init is the number of times the algorithm will run with different centroids.
# The final results will be the best output of n_init runs in terms of inertia.
kmeans = KMeans(n_clusters=5, n_init=10, random_state=0)
df['cluster'] = kmeans.fit_predict(df[zscaled_columns])

# Display mean feature values per cluster for interpretation.
# Helpful for understanding the nature of each cluster.
for cluster in range(5):
    print(f"Cluster {cluster}")
    print(df[df['cluster'] == cluster][zscaled_columns].mean())
    #visual separator -----
    print('-'*40)
    
# Save the modified DataFrame to a new CSV file
df.to_csv('C:\\Users\\gothu\\Downloads\\FinalFulldata.csv', index=False)
```

## Elbow Plot
Before proceeding with our freshly obtained cluster data, it's crucial to determine the appropriate number of clusters relative to the number of metrics we have. To achieve this, I utilized an elbow plot to assess the relevance of different cluster numbers.

We then visualized the plot and concluded that the selected number of clusters aligns with the results.

```python
# Elbow method to determine optimal number of clusters. 
# Plotting inertia against the number of clusters can provide insight into how many clusters are appropriate.
def plot_elbow_curve(model, data, cluster_ranges):
    inertia_values = []
    for i in cluster_ranges:
        model.set_params(n_clusters=i)
        model.fit(data)
        inertia_values.append(model.inertia_)

    plt.figure(figsize=(8.27,5.87))
    plt.plot(cluster_ranges, inertia_values, marker='o')
    plt.title("Elbow Curve")
    plt.xlabel("Number of clusters")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()


plot_elbow_curve(kmeans, df[zscaled_columns], cluster_ranges=range(1, 15))
```

##  Correlation Matrix Heatmap


We now need to ensure that our metrics aren't highly correlated. Highly correlated metrics can introduce confusion and challenges for the k-means clustering algorithm, potentially leading to suboptimal results.

To address this concern, I am using a correlation matrix complemented by a heatmap. This visualization technique allows us to explore the degree of correlation between different metrics. By identifying and understanding any significant correlations, we can make informed decisions about feature selection and ensure that the clustering process is based on relevant and independent attributes.

This step is crucial as it helps us avoid issues that could arise from redundant or overly related metrics during the clustering process.

This marks the completion of the data transformation and aggregation tasks required for the clustering algorithm, as well as its .execution

```python
# Validate Correlations for better feature usage in clusters.
# Load the data into a pandas DataFrame
# This CSV contains z-scaled features for k-means clustering
correlation_df=pd.read_csv('C:\\Users\\gothu\\Downloads\\FinalFulldata.csv')

# Compute the correlation matrix for a subset of columns
# These are potential features for k-means clustering
# List of features to be set inside correlation_matrix (all are z-scaled)
correlation_matrix = correlation_df[[ 'total_cancellation_rate_zscaled',
                          'average_checked_bags_zscaled', 'conversion_rate_zscaled',
                          'prefers_flights_zscaled', 'prefers_hotels_zscaled', 'prefers_both_zscaled',
                          'discount_responsiveness_zscaled',
                          'flight_discount_proportion_zscaled', 'average_flight_discount_zscaled',
                          'hotel_discount_proportion_zscaled', 'average_hotel_discount_zscaled', 'both_discount_proportion_zscaled',
                          'flight_hunter_index_zscaled', 'hotel_hunter_index_zscaled',
                          'total_hotel_usd_spent_zscaled', 'total_flight_usd_spent_zscaled'
                          ]].corr()

# Generate a heatmap for better visual inspection of the correlation matrix
plt.figure(figsize=(16.54,11.74))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Metrics')
plt.xticks(rotation=45, ha='right')

# Save the heatmap to a file for future reference or sharing
plt.savefig('C:\\Users\\gothu\\Downloads\\CorrelationVerification.png', dpi=96, bbox_inches='tight')
plt.show()

# Initialize set to hold pairs to be dropped (upper triangle of the matrix)
# This avoids redundancy in the correlation matrix as it is symmetric
pairs_to_drop = set()
cols = correlation_matrix.columns
for i in range(0, correlation_matrix.shape[1]):
    for j in range(0, i+1):
        pairs_to_drop.add((cols[i], cols[j]))

# Unstack the correlation matrix and sort by absolute value
# The 'abs' function converts all correlation coefficients to their absolute values
corr_abs_unstacked = correlation_matrix.abs().unstack()

# Drop pairs to get unique (i, j) correlations (avoid redundancy)
# Sort remaining pairs by descending correlation
corr_abs_unstacked = corr_abs_unstacked.drop(labels=pairs_to_drop).sort_values(ascending=False)

# Extract highly correlated pairs (absolute value greater than 0.8)
# These are potential candidates for feature reduction or removal
high_corr = corr_abs_unstacked[corr_abs_unstacked > 0.8]

# Display pairs of highly correlated features
print("Highly correlated pairs: ")
print(high_corr)
# Note that features with high correlation may need to be treated carefully during modeling
print("\nThanks to this we can now remove the highly correlated values")
```

## Cluster Heatmap

After ensuring the adequacy of our dataset for clustering in the previous steps, we can now proceed to work with and visualize the clusters we've obtained.

I have created a heatmap to visualize the clusters and labeled them based on their respective "Tribe" behuavior. The labeling process takes into consideration their metrics results, which you can observe in the cluster heatmap below.

This step is essential as it forms the basis for assigning perks to each "Tribe," allowing us to tailor our offerings effectively.

```python
# Create a heatmap for cluster characteristics.
# Allows for easier interpretation of scaled features across clusters.

cluster_summaries = {}  # A dictionary to store the mean value DataFrame of each cluster

for cluster in range(5):  # Loop over all clusters (you specified 5 clusters in your KMeans)
    sub_df = df[df['cluster'] == cluster]
    cluster_summaries[cluster] = sub_df[zscaled_columns].mean()

# Step 2: Convert the dictionary into a DataFrame for visualization
# Here, each row corresponds to a cluster and each column to a feature

cluster_summary_df = pd.DataFrame.from_dict(cluster_summaries, orient='index', columns=zscaled_columns)
# Swap the placement of the x and y axis.
cluster_summary_transposed = cluster_summary_df.T
# Define custom x-axis labels
custom_x_labels= ['Discount Enthusiasts', 'Balanced Travellers', 'Hotel Discount Lovers', 'Jet-Set & Indecisive', 'Hotel Loyalists']

# Set the custom labels to the DataFrame
cluster_summary_transposed.columns = custom_x_labels

# Step 3: Create heatmap
plt.figure(figsize=(8.27,5.87))

sns.heatmap(cluster_summary_transposed, annot=True, cmap="coolwarm", annot_kws={"size":10}, fmt='.3f')
plt.title('Traveller Tribes Heatmap')
plt.ylabel('Features')
plt.xticks(rotation=45, ha='right')
plt.savefig('C:\\Users\\gothu\\Downloads\\final_kmeans_clustering_heatmap.png', dpi=96, bbox_inches='tight')
plt.show()
```
